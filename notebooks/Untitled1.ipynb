```python
import numpy as np

# Load the .npy file from the specified path
file_path = '/Volumes/ARCHIVES/thesis_pipeline/hmm_segment_data/TD_final_events.npy'
optimal_k_data = np.load(file_path, allow_pickle=True)

# Print the data type to inspect if it's a dictionary or list
print(f"Type of data: {type(optimal_k_data)}")

# Simply print the content (if it's not too large) or explore its structure
print(optimal_k_data)  # Without indexing

```

```python
# If it's a dictionary, print the first few entries
if isinstance(optimal_k_data, dict):
    for i, (key, value) in enumerate(optimal_k_data.items()):
        print(f"Region: {key}, Optimal k: {value}")
        if i >= 5:  # Print only first 5 entries for brevity
            break
else:
    # If it's an array or other data type, just print it to inspect
    print(optimal_k_data)
```

```python
# Extract the dictionary from the 'model' key
model_data = optimal_k_data['model']

# Extract and print optimal k values (n_events) for each region
optimal_k_values = {}
for region, event_segment in model_data.items():
    optimal_k = event_segment.n_events  # Extract the number of events (k)
    optimal_k_values[region] = optimal_k
    print(f"Region: {region}, Optimal k: {optimal_k}")
```
